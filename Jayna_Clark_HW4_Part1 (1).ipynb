**Q1** Please use Pandas to read olympic_medals.csv and use parallel_categories function from plotly.express to visualize proportions of medal type for each gender from since year 2000. Please see the example in the Python notebook we walked through in the class.

```python
from google.colab import drive
drive.mount('/content/drive')

olympic = pd.read_csv('/content/drive/MyDrive/olympic_medals.csv')


#filter only years after/= 2000
olympic = olympic[olympic['Year']>=2000]
```

```python
#Please use this cell to create your your figure. Please use Year column to color your graph.

import plotly.express as px
px.parallel_categories(olympic,
    dimensions=['Year', 'Medal', 'Gender'],
    color=olympic['Year'].astype(int),
    color_continuous_scale=px.colors.sequential.Plasma
)
```

Make sure that your x tick labels are those four states above.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# create DataFrame
df = pd.DataFrame({'High Temp': [28, 30, 34, 38, 45, 42,
                                 38, 35, 32, 28, 25, 21],
                   'Low Temp': [22, 26, 30, 32, 41, 38,
                                32, 31, 28, 22, 15, 15],
                   'Avg Temp': [25, 28, 32, 35, 43, 40,
                                35, 33, 30, 25, 20, 18]},
                  index=['Jan', 'Feb', 'Mar', 'Apr', 'May',
                         'Jun', 'Jul', 'Aug', 'Sep', 'Oct',
                         'Nov', 'Dec'])


# create stacked bar chart for monthly temperatures
df.plot(kind='bar', stacked=True, color=['red', 'skyblue', 'green'])

# labels for x & y axis
plt.xlabel('Months')
plt.ylabel('Temp ranges in Degree Celsius')

# title of plot
plt.title('Monthly Temperatures in a year')
```

```python
#You can use this cell to write your code. It is doable at most 4 lines of code.

election = pd.read_csv('/content/drive/MyDrive/2016elections.csv')
election = election[election['st'].isin(['AR', 'MI', 'CA', 'WI'])]
vote_pct_columns = ['pct_clinton', 'pct_trump', 'pct_johnson', 'pct_other']

election[vote_pct_columns].plot(kind='bar', stacked=True, color=['blue', 'red', 'green','grey'])
plt.xlabel('State')
plt.ylabel('Percent Votes for Candidate')
plt.title('Percent Votes for Candidates in 2016 Election by Select States')
plt.xticks(range(len(election['st'])), election['st'])
plt.legend(['Clinton', 'Trump', 'Johnson', 'Other'])
```
